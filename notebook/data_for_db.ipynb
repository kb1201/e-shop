```python
import pandas as pd

df = pd.read_csv("amazon_with_img_links_and_names.csv")
```

```python
columns = [
     'product_name', 'category', 'discounted_price', 'actual_price',
    'discount_percentage', 'rating', 'rating_count', 'about_product',
    'img_link', 'product_link', 'specific_category', 'combined_text',
    'image_name', 'image_path'
]
```

```python
df = df[columns]
df.head()
```

```python
df = df.rename(columns={
    "product_name": "name",
})
```

```python
print(df.dtypes)
```

```python
df.head()
```

```python
def safe_convert_to_int(x):
    try:
        return int(float(x)) if not pd.isna(x) else pd.NA
    except (ValueError, TypeError):
        return pd.NA

df['rating_count'] = df['rating_count'].apply(safe_convert_to_int).astype('Int64')

# Verify conversion
print("Rating count sample:")
print(df[['rating_count']].head(10))
print("\nNull counts:")
print(df['rating_count'].isna().sum())
```
