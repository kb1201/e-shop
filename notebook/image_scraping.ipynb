```python
def get_unsplash_image_url(query, api_key):
    url = f'https://api.unsplash.com/photos/random?query={query}&client_id={api_key}&count=1'
    response = requests.get(url)

    if response.status_code == 200:
        image_data = response.json()
        if image_data:
            return image_data[0]['urls']['regular']  # URL of the image

    print(response.status_code)
    return None

unsplash_api_key = "key"
```

```python
import time

def get_pixabay_image_url(query, api_key):
    url = f'https://pixabay.com/api/?key={api_key}&q={query}&image_type=photo&pretty=true'
    response = requests.get(url)
    time.sleep(1)

    if response.status_code == 200:
        image_data = response.json()
        print(image_data)
        if image_data['hits']:
            return image_data['hits'][0]['webformatURL']  # URL of the image

    print(response.status_code)
    return None

pixabay_api_key = 'key'
```

```python
import requests
import pandas as pd

def get_pexels_image(query, api_key):
    url = f'https://api.pexels.com/v1/search'
    params = {
        'query': query,
        'per_page': 1
    }
    headers = {
        'Authorization': api_key
    }

    response = requests.get(url, headers=headers, params=params)

    if response.status_code == 200:
        data = response.json()
        if data['photos']:
            return data['photos'][0]['url']

    print(response.status_code)
    return None


def update_img_links_for_none(df, api_key, get_image, query_func):
    for idx, row in df.iterrows():
        if pd.isna(row['img_link']):

            query = query_func(row)
            img_url = get_image(query, api_key)
            if img_url:
                df.at[idx, 'img_link'] = img_url
    return df


pexels_api_key = 'key'

def create_query_1(row):
    query = row['product_name'] + ' ' + row['specific_category']
    return query

def create_query_2(row):
    query = row['specific_category']
    return query

```

```python
df = update_img_links_for_none(df, pexels_api_key, get_pexels_image, create_query_1)
print(df)
```

```python
df = update_img_links_for_none(df, unsplash_api_key, get_unsplash_image_url, create_query_1)
print(df)
```

## Download images to drive and add img_path to data

```python
import os
import requests
from PIL import Image
from io import BytesIO
import re
from bs4 import BeautifulSoup
```

```python
def sanitize_filename(filename):
    """Zamenjuje nedozvoljene karaktere u imenu fajla."""
    filename = filename.replace("\xa0", " ")  # Zamena non-breaking space
    filename = re.sub(r'[\\/*?:"<>|]', "", filename)  # Uklanja zabranjene karaktere
    filename = filename.replace(" ", "_")  # Zamenjuje space sa "_"
    return filename[:100]  # Skrati ime ako je predugo
```

```python
drive_folder = '/content/drive/My Drive/e_shop_images'
os.makedirs(drive_folder, exist_ok=True)

if 'image_name' not in df.columns:
    df['image_name'] = None
if 'image_path' not in df.columns:
    df['image_path'] = None

def get_pexels_image(photo_id):
    """Fetches the direct image URL using the Pexels API."""
    url = f"https://api.pexels.com/v1/photos/{photo_id}"
    headers = {"Authorization": pexels_api_key}

    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        data = response.json()
        return requests.get(data["src"]["original"], headers=headers)  # Get the original image URL

    print(f"Failed to fetch image from Pexels API, Status: {response.status_code}")
    return None

img = get_pexels_image('19153798')
img = Image.open(BytesIO(img.content))
```

```python
def extract_photo_id(url):
    match = re.search(r'/photo/.*?(\d+)/$', url)
    return match.group(1) if match else None
```

```python


def download_and_save_image(image_url, save_dir, product_name):
    """Downloads an image, handling Pexels links by extracting the direct image URL first."""
    try:
        # If the URL is from Pexels, extract the direct image URL
        headers = {
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
        }

        if "pexels.com" in image_url:
            print(f"Extracting image URL from Pexels page: {image_url}")
            photo_id = extract_photo_id(image_url)
            response = get_pexels_image(photo_id)
            if not image_url:
                return None, None
        else:
          response = requests.get(image_url, headers=headers, stream=True)

        if response.status_code == 200:
            clean_product_name = sanitize_filename(product_name)
            img_name = f"{clean_product_name}.jpg"
            save_path = os.path.join(save_dir, img_name)

            img = Image.open(BytesIO(response.content))

            if img.mode == "RGBA":
                img = img.convert("RGB")

            img.save(save_path, "JPEG")
            return img_name, save_path
        else:
            print(f"Failed to download {image_url}, Status: {response.status_code}")
            return None, None
    except Exception as e:
        print(f"Error downloading {image_url}: {e}")
        return None, None

for index, row in df.iterrows():
    img_url = row['img_link']
    product_name = row['product_name']

    if pd.notna(img_url):
        img_name, img_path = download_and_save_image(img_url, drive_folder, product_name)

        if img_name and img_path:
            df.at[index, 'image_name'] = img_name
            df.at[index, 'image_path'] = img_path

df.to_csv('/content/drive/My Drive/e_shop_data/amazon_with_img_links_and_names.csv', index=False)

print("Sve slike su preuzete i sačuvane u Google Drive-u!")
```
